```python
import numpy as np 
import matplotlib.pyplot as plt 
import random

def random_walk():
    initial_point = 0.0
    position = initial_point 
    num_steps = 1000
    positions = [position]

    for _ in range(num_steps): 
        direction = random.choice([-1, 1]) 
        position += direction 
        positions.append(position)
    return positions 
positions = random_walk() 

plt.figure(figsize = (16, 6))
plt.plot(range(len(positions)), positions)
plt.xlabel('Step')
plt.ylabel('Position')
plt.title('One-Dimensional Random Walk Trajectory')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class Random_walk: 
    def __init__(self, i_position, num_steps):
        self.i_position = i_position 
        self.num_steps = num_steps
    
    def random_w(self): 
        positions = [self.i_position]
        for _ in range(self.num_steps):
            direction = np.random.choice([-1, 1]) 
            next_position = positions[-1] + direction 
            positions.append(next_position) 
        return positions 

random_walker = Random_walk(np.random.uniform(-1, 1), 2000)  
positions = random_walker.random_w() 

plt.figure(figsize=(16, 6))
plt.plot(range(len(positions)), positions)
plt.xlabel('Step')
plt.ylabel('Position')
plt.title('One-Dimensional Random Walk Trajectory')
plt.grid(True)
plt.show()
```

```python
# 2D random walk
import numpy as np 
import random 
import matplotlib.pyplot as plt 

class RandomWalk: 
    def __init__(self, num_steps): 
        self.num_steps = num_steps 
        self.position = []

    def random(self, x_range, y_range): 
        initial_position = [random.uniform(x_range[0], x_range[1]), random.uniform(y_range[0], y_range[1])] 
        return initial_position 
    
    def simulate_walk(self, initial_position): 
        current_position = initial_position
        directions = [(0, 1), (0, -1), (-1, 0), (1, 0)]

        self.position.append(initial_position) 
        
        for _ in range(self.num_steps):
            direction = random.choice(directions)
            current_position = [current_position[0] + direction[0], current_position[1] + direction[1]]
            self.position.append(current_position)  
        
        return self.position  

num_steps = 3000
x_range = (-10, 10)
y_range = (-10, 10)

random_walker = RandomWalk(num_steps)
initial_position = random_walker.random(x_range, y_range)
positions = random_walker.simulate_walk(initial_position)

positions = np.array(positions)

plt.figure(figsize=(12, 6))
plt.plot(positions[:, 0], positions[:, 1])
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.title('Two-Dimensional Random Walk Trajectory')
plt.grid(True)
plt.show()
```
